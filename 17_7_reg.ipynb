```python
import sys
sys.path.append('../')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Function
from torch.nn import Module

from low_rank_rnns.modules import *
from low_rank_rnns.my_low_rank import *
from low_rank_rnns import rdm, helpers, stats

import warnings
warnings.filterwarnings('ignore')

size = 512
noise_std = 5e-2
alpha = .2
```

```python
# generate traning data for random dot motion task 
x_train, y_train, mask_train, x_val, y_val, mask_val = rdm.generate_rdm_data(1000)

# train a 1-rank RNN to solve the task with smart activation 

teacher_net = LowRankRNN(
                        input_size=1,
                        hidden_size=512,
                        output_size=1,
                        noise_std=noise_std,
                        alpha=alpha,
                        rank=1,
                        )

# train the net 
train(
  net=teacher_net,
  _input= x_train,
  _target=y_train,
  _mask=mask_train,
  n_epochs=1_000,
  lr=1e-4,
  keep_best=True,
  plot_learning_curve=True,
  cuda=True,
  smart=False,
)

torch.save(teacher_net.state_dict(), f'17_7/teacher_net_rank_1_tanh.pt')

teacher_net.to(device='cpu')

error, accuracy = rdm.test_rdm(net=teacher_net, x=x_val, y=y_val, mask=mask_val)

print('Error:', error)
print('Accuracy:', accuracy)

rdm.psychometric_curve_rdm(teacher_net)
plt.show()

rdm.psychometric_matrix(teacher_net)
plt.show()

rdm.plot_outputs(teacher_net)
plt.show()
```

```python
# generate trajectories 
output, traj = teacher_net.forward(x_train, return_dynamics=True)
n_trials = x_train.shape[0]
T = x_train.shape[1]
target = torch.tanh(traj[:,1:].detach())

out1, traj1 = teacher_net.forward(x_val, return_dynamics=True)
traj1 = torch.tanh(traj1)
y1 = traj1.detach().numpy().ravel()

all_r2, all_acc = [],[]

for xx in range(4):
    
    list_of_r2 = []
    list_of_acc = []

    # fit increasing rank of RNN to teacher-RNN trajectories
    for rank in [1,2,3,4,5]:

        print(f'fitting rank: {rank}')

        student_net = LowRankRNN(
                        input_size=1,
                        hidden_size=size,
                        output_size=size,
                        noise_std=0, # noise_std 
                        alpha=alpha, 
                        rank=rank, 
                        wo_init= size * torch.from_numpy(np.eye(size)), 
                        train_wi=True,
                        train_so=False,
                        )

        train(
            net=student_net,
            _input=x_train,
            _target=target,
            _mask=torch.ones((n_trials, T, 1)), 
            n_epochs=500, 
            lr=1e-2,
            clip_gradient=1.,
            keep_best=True,
            cuda=True,
            smart=False
        )

        torch.save(student_net.state_dict(), f'17_7/student_net_rank_{rank}_tanh_{xx}.pt')
        print()

        student_net.to(device='cpu')

        out2, traj2 = student_net.forward(x_val, return_dynamics=True)
        traj2 = student_net.non_linearity(traj2)
        y2 = traj2.detach().numpy().ravel()

        r2 = stats.r2_score(y1, y2)
        print('r2: ',r2)
        list_of_r2.append(r2)

        # Replace output identity matrix by output vector and compute task performance
        student_net.wo = nn.Parameter(teacher_net.wo_full.clone())
        student_net.output_non_linearity = torch.tanh
        student_net.output_size = 1
        student_net.so = nn.Parameter(torch.tensor([1. * size]))
        
        error, accuracy = rdm.test_rdm(net=student_net, x=x_val, y=y_val, mask=mask_val)    
        print('accuracy: ',accuracy)
        list_of_acc.append(accuracy)

        print('======================')
        
    all_r2.append(list_of_r2)
    all_acc.append(list_of_acc)
    
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(7,3))

r2 = np.array(all_r2)
sns.pointplot(ax=ax1, data=r2, errorbar='se', color='indianred', markers='o',)
ax1.set_ylim(0,1.1)
ax1.set_xlabel('Rank',size=14)
ax1.set_ylabel('$R^2$',size=14)
ax1.set_xticks([0, 1, 2, 3, 4])
ax1.set_xticklabels([1, 2, 3, 4, 5])

acc = np.array(all_acc)
sns.pointplot(ax=ax2, data=acc, errorbar='se', color='indianred', markers='o',)

ax2.set_ylim(0,1.1)
ax2.set_xlabel('Rank',size=14)
ax2.set_ylabel('Task accuracy',size=14)
ax2.set_xticks([0, 1, 2, 3, 4])
ax2.set_xticklabels([1, 2, 3, 4, 5])

sns.despine()
plt.tight_layout()
plt.savefig('../tanh.pdf')
plt.show()
```
